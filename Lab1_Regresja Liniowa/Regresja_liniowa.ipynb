```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import scipy
import time
import pandas as pd 
import numpy as np

times = 100

x, Y  = datasets.make_regression(n_features=1, noise = 16)
poly = PolynomialFeatures(1)
X = poly.fit_transform(x)

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state = 244828)
```

### Metoda Numeryczna 
### $$1/N(y-X^T*w)^T(y-X^T*w)$$

```python
def func(w, X, Y):
    xTw = np.matmul(X, w)
    y_sub_xTw = np.subtract(Y, xTw)
    return np.matmul(np.transpose(y_sub_xTw), y_sub_xTw)*(1/len(X))

start_time_num = time.time()
for i in range(times):
    ans_num = scipy.optimize.minimize(func, [1,1], args = (x_train, y_train), method='Powell')
end_time_num = (time.time() - start_time_num)/times

lst_num = ans_num.x
alpha_num = lst_num[1]
beta_num = lst_num[0]

print(lst_num)
```

```python
def func(w, X, Y):
    yxtans = np.subtract(np.reshape(Y, (len(Y), 1)), np.matmul(X, np.transpose(np.reshape(w,(1,2)))))
    yxtansyxtans = np.matmul(np.transpose(yxtans),yxtans)
    return yxtansyxtans*(1/len(x_train))

start_time_num = time.time()
for i in range(times):
    ans_num = scipy.optimize.minimize(func, [1,1], args = (x_train, y_train), method='Powell')
end_time_num = (time.time() - start_time_num)/times

lst_num = ans_num.x[0]
alpha_num = lst_num[1]
beta_num = lst_num[0]

print(lst_num)
```

### Metoda analityczna 
###  $$ (X^T*X)^{-1}*X^T*Y$$

```python
start_time_ana = time.time()
for i in range(times):
    alpha_beta = np.matmul(np.linalg.inv(np.matmul(np.transpose(x_train),x_train)),np.matmul(np.transpose(x_train),y_train))
end_time_ana = (time.time() - start_time_ana)/times

alpha_ana = alpha_beta[1]
beta_ana = alpha_beta[0]
lst_ana = [alpha_ana, beta_ana]

print(lst_ana)
```

### Wykresy

```python
x_train = x_train.flatten()
x_test = x_test.flatten()


plt.figure(figsize=(24,12))
plt.scatter(x_train[1::2], y_train, color='red', alpha=0.5, marker = '+', label="Train")
plt.scatter(x_test[1::2], y_test, color='blue', alpha=0.5, marker = '+', label="Test")
plt.plot([min(x), max(x)], [alpha_num*min(x) + beta_num, alpha_num*max(x) + beta_num], label = "Numeryczna")
plt.plot([min(x), max(x)], [alpha_ana*min(x) + beta_ana, alpha_ana*max(x) + beta_ana], label = "Analityczna")
plt.legend()
plt.show
```

### MSE

```python
xPlot = x_test[1::2]
yPredNum = np.empty(len(y_test))
yPredAna = np.empty(len(y_test))

# Numeryczna
for i in range(len(y_test)):
    yPredNum[i] = alpha_num*xPlot[i]+beta_num
mse_num = mean_squared_error(y_test, yPredNum)

# Analityczna
for i in range(len(y_test)):
    yPredAna[i] = alpha_ana*xPlot[i]+beta_ana
mse_ana = mean_squared_error(y_test, yPredAna)
```
